```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/employee_data_250-2.csv')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Salary'], bins=10, kde=True, color='blue')
plt.xlabel("Salary (USD)")
plt.ylabel("Number of Employees")
plt.title("Salary Distribution")
plt.show()
```

```python
salary_range = pd.cut(df['Salary'], bins=10).value_counts()
print("Most common salary range:", salary_range.idxmax())
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Experience (Years)"], y=df["Salary"], color="blue", alpha=0.6)
sns.regplot(x=df["Experience (Years)"], y=df["Salary"], scatter=False, color="red")

plt.xlabel("Experience (Years)")
plt.ylabel("Salary (USD)")
plt.title("Experience vs. Salary")
plt.show()

# Calculate correlation between Experience and Salary
correlation = df["Experience (Years)"].corr(df["Salary"])
print(f"Correlation between Experience and Salary: {correlation}")
```

```python
plt.figure(figsize=(8, 5))
department_counts = df['Department'].value_counts()
sns.barplot(x=department_counts.index, y=department_counts.values, palette='viridis')
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees per Department")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Job Satisfaction', data=df, palette='muted')
plt.xlabel("Department")
plt.ylabel("Count of Employees")
plt.title("Job Satisfaction Distribution Across Departments")
plt.legend(title='Job Satisfaction')
plt.xticks(rotation=45)
plt.show()
```
